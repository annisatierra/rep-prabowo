```python
import os
import pandas as pd
import glob

file_list = glob.glob('rep*.xlsx')
df_list = []

for file in file_list:
    try:
        # Attempt to read each file
        df = pd.read_excel(file)
        if not df.empty:  # Check if DataFrame has any data
            # Menambahkan kolom baru dengan nama file tanpa ekstensi .xlsx
            file_name = os.path.splitext(os.path.basename(file))[0]
            df['FileName'] = file_name
            
            # Menempatkan kolom 'FileName' sebagai kolom pertama
            cols = ['FileName'] + [col for col in df.columns if col != 'FileName']
            df = df[cols]

            # Menambahkan DataFrame ke daftar
            df_list.append(df)
    except pd.errors.EmptyDataError:
        print(f"Skipping empty file: {file}")

# Concatenate all DataFrames if any were loaded
if df_list:
    combined_df1 = pd.concat(df_list, ignore_index=True)
else:
    print("No valid data files found to combine.")
```

```python
import os
import pandas as pd
import glob

file_list = glob.glob('185*.xlsx')
df_list = []

for file in file_list:
    try:
        # Attempt to read each file
        df = pd.read_excel(file)
        if not df.empty:  # Check if DataFrame has any data
            # Menambahkan kolom baru dengan nama file tanpa ekstensi .xlsx
            file_name = os.path.splitext(os.path.basename(file))[0]
            df['FileName'] = file_name

            # Menempatkan kolom 'FileName' sebagai kolom pertama
            cols = ['FileName'] + [col for col in df.columns if col != 'FileName']
            df = df[cols]

            # Menambahkan DataFrame ke daftar
            df_list.append(df)
    except pd.errors.EmptyDataError:
        print(f"Skipping empty file: {file}")

# Concatenate all DataFrames if any were loaded
if df_list:
    combined_df2 = pd.concat(df_list, ignore_index=True)
else:
    print("No valid data files found to combine.")
```

```python
df_all = pd.concat([combined_df1, combined_df2], ignore_index=True)
df_all
```
